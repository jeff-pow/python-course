```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
# Create the x and y indicies
x = [x for x in range(-50, 51, 1)]
x = np.array([i / 100 for i in x])
y = [y for y in range(-50, 51, 1)]
y = np.array([i / 100 for i in y])
```

```python
# Turn the indicies into a meshgrid
xv, yv = np.meshgrid(x, y, indexing='xy')
```

```python
# Create variables to hold locations for the two particles
p1x = .105
p1y = .105
p2x = .205
p2y = -.205
```

```python
# Create the function that evaluates the grid with the first particle on it
grid1 = 1 / 9e9 * 1e-9 / np.sqrt((p1x - xv) ** 2 + (p1y - yv) ** 2)
```

```python
# Plot the grid
plt.contourf(xv, yv, grid1)
plt.grid()
plt.axis('scaled')
plt.title('Electric Potential Due to 1 nC Charge')
plt.ylabel('Displacement (m)')
plt.xlabel('Displacement (m)')
plt.colorbar()
plt.show()
```

```python
# Create the function that evaluates the grid with the second particle on it
grid2 = 1 / 9e9 * 1e-9 * 2 / np.sqrt((p2x - xv) ** 2 + (p2y - yv) ** 2)
```

```python
# Plot the grid
plt.contourf(xv, yv, grid2)
plt.grid()
plt.axis('scaled')
plt.title('Electric Potential Due to 2 nC Charge')
plt.ylabel('Displacement (m)')
plt.xlabel('Displacement (m)')
plt.colorbar()
plt.show()
```

```python
# Create the function that evaluates the grid with both particles on it
grid3 = 1 / 9e9 * 1e-9 * (1 / np.sqrt((p1x - xv) ** 2 + (p1y - yv) ** 2) + 2 / np.sqrt((p2x - xv) ** 2 + (p2y - yv) ** 2))
# Plot the grid
plt.contourf(xv, yv, grid3)
plt.grid()
plt.axis('scaled')
plt.title('Electric Potential Due to Both Charges')
plt.ylabel('Displacement (m)')
plt.xlabel('Displacement (m)')
plt.colorbar()
plt.scatter(0, 0, color='red')
plt.show()
```

```python
# Evaluate the grid with two particles on it for the electric potential at the origin
print(grid3[0][0])
# Electric potential at the origin is evaluated at 4.21e-19 volts
```
