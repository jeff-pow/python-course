```python
import pandas as pd
import geopandas as geo
from geodatasets import get_path
import matplotlib.pyplot as plt
```

```python
df = pd.DataFrame(
    {
        "City": ["Chapel Hill", "Houston", "Gainesville", "Berlin", "Reykjavik"],
        "Country": ["USA", "USA", "USA", "Germany", "Iceland"],
        "Latitude": ["35.9132", "29.7604", "29.6520", "52.52", "64.147"],
        "Longitude": ["-79.0558", "-95.3698", "-82.3250", "13.4050", "21.9408"],
    }
)
df.head()
```

```python
gdf = geo.GeoDataFrame(
    df, geometry=geo.points_from_xy(df.Longitude, df.Latitude), crs="EPSG:4326"
)
gdf.plot()
```

```python
world = geo.read_file(get_path("naturalearth.land"))
```

```python
ax = world.clip([-200, -200, 200, 200]).plot(color="white", edgecolor="black")
gdf.plot(ax=ax, color="red")
plt.show()
```
