```python
import numpy as np
import pandas as pd
pd.options.mode.copy_on_write = True
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.formula.api as smf
from sklearn import linear_model
```

```python
porsche = pd.read_csv('Porsche.csv')
jaguar = pd.read_csv('Jaguar.csv')
```

```python
smf.ols(formula = 'Price ~ Mileage + Age', data=porsche).fit().summary()
```

```python
smf.ols(formula = 'Price ~ Mileage', data=porsche).fit().summary()
```

```python
smf.ols(formula = 'Price ~ Age', data=porsche).fit().summary()
```

```python
smf.ols(formula = 'Price ~ Mileage + Age', data=jaguar).fit().summary()
```

```python
smf.ols(formula = 'Price ~ Mileage', data=jaguar).fit().summary()
```

```python
smf.ols(formula = 'Price ~ Age', data=jaguar).fit().summary()
```

```python
jaguar['cartype'] = 'jaguar'
porsche['cartype'] = 'porsche'
total = pd.concat([jaguar, porsche], ignore_index=True)
```

```python
smf.ols(formula = 'Price ~ Mileage + Age', data=total).fit().summary()
```

```python
smf.ols(formula = 'Price ~ Mileage', data=total).fit().summary()
```

```python
smf.ols(formula = 'Price ~ Age', data=total).fit().summary()
```

```python
lr = linear_model.LinearRegression()
predicted = lr.fit(X=porsche[['Age', 'Mileage']], y=porsche['Price'])
value = predicted.coef_[0] * 10 + predicted.coef_[1] * 55 + predicted.intercept_
print(predicted.coef_[0], predicted.coef_[1], predicted.intercept_)
# Expected value of a porsche 10 years old with 55k miles
value *= 1000
value
```

```python
lr = linear_model.LinearRegression()
predicted = lr.fit(X=jaguar[['Age', 'Mileage']], y=jaguar['Price'])
value = predicted.coef_[0] * 10 + predicted.coef_[1] * 55 + predicted.intercept_
print(predicted.coef_[0], predicted.coef_[1], predicted.intercept_)
# Expected value of a jaguar 10 years old with 55k miles
value *= 1000
value
```

```python
lr = linear_model.LinearRegression()
predicted = lr.fit(X=total[['Age', 'Mileage']], y=total['Price'])
value = predicted.coef_[0] * 10 + predicted.coef_[1] * 55 + predicted.intercept_
print(predicted.coef_[0], predicted.coef_[1], predicted.intercept_)
# Expected value of a sports car 10 years old with 55k miles
value *= 1000
value
```

```python
fig, axes = plt.subplots(1, 3, figsize=(15,5))
age = total['Age']
mileage = total['Mileage']
price = total['Price']
# Code I learned about 3D plots from: 
# https://matplotlib.org/stable/gallery/mplot3d/scatter3d.html#sphx-glr-gallery-mplot3d-scatter3d-py
axes[0] = fig.add_subplot(131, projection='3d')
axes[0].scatter(age, mileage, price, c='b', marker='o')


sns.regplot(x=age, y=price, ax=axes[0], ci=None, color='r')
sns.regplot(x=mileage, y=price, ax=axes[0], ci=None, color='g')
axes[0].view_init(elev=25, azim=10)

sns.regplot(x='Mileage', y='Price', data=total, ci=None, ax=axes[1])
sns.regplot(x='Age', y='Price', data=total, ci=None, ax=axes[2])
axes[0].set_xlabel('Age')
axes[0].set_ylabel('Mileage')
axes[0].set_zlabel('Price')

plt.show()
# Couldn't figure out how to get rid of the 2D axes under the 3D graph unfortunately...
```
